```
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import plotly.express as px
```

```
plt.rcParams['figure.figsize']=10,12

import warnings
warnings.filterwarnings('ignore')
```

```
url4='https://raw.githubusercontent.com/khushbu1998/covid19/master/Russia_23rdMay.csv'
```

```
Russia= pd.read_csv(url4)
```

```
Russia
```

```
Russia['Date '] = pd.to_datetime(Russia['Date '])
Russia= Russia.sort_values('Date ')
```

```
Russia.style.background_gradient(cmap='Reds')
```

```
train_dataset= pd.DataFrame()
train_dataset['ds']=Russia['Date ']
train_dataset['y']=Russia['Cases']
train_dataset
```

```
from fbprophet import Prophet
from fbprophet.plot import plot_plotly
m = Prophet(interval_width=0.95)
m.fit(train_dataset)
future = m.make_future_dataframe(periods=130)
```

```
forecast=m.predict(future)
```

```
print("Cases per day prediction: Russia")
fig1 = m.plot(forecast,xlabel="Date",ylabel="Cases")
```

```
from fbprophet.diagnostics import cross_validation
cv=cross_validation(m,horizon='40 days',period='10 days',initial='10 days')
```

```
cv
```

```
from fbprophet.diagnostics import performance_metrics
model=performance_metrics(cv)
model.tail()
```

```
from fbprophet.plot import plot_cross_validation_metric
plot_cross_validation_metric(cv,metric='rmse')
```
