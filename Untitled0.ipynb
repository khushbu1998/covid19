```
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import plotly.express as px
import plotly.graph_objects as go
```

```
url1 ='https://raw.githubusercontent.com/khushbu1998/covid19/master/Italy_17thmay.csv'
url2='https://raw.githubusercontent.com/khushbu1998/covid19/master/Brazil_17thmay.csv'
url3='https://raw.githubusercontent.com/khushbu1998/covid19/master/India_17thmay.csv'
url4='https://raw.githubusercontent.com/khushbu1998/covid19/master/Russia_17thmay.csv'
```

```
Italy = pd.read_csv(url1)
Brazil = pd.read_csv(url2)
India= pd.read_csv(url3)
Russia= pd.read_csv(url4)
```

```
Italy['Date '] = pd.to_datetime(Italy['Date '])
Italy= Italy.sort_values('Date ')
Brazil['Date '] = pd.to_datetime(Brazil['Date '])
Brazil= Brazil.sort_values('Date ')
India['Date '] = pd.to_datetime(India['Date '])
India= India.sort_values('Date ')
Russia['Date '] = pd.to_datetime(Russia['Date '])
Russia= Russia.sort_values('Date ')
```

```
Italy
```

```
from plotly.subplots import make_subplots
```

```
fig=make_subplots(rows=2,cols=2,specs=[[{},{}],[{},{}]], subplot_titles=('Italy','Brazil','India','Russia'))
```

```
fig.add_trace(go.Bar(x=Italy['Date '],y=Italy['Cummulative cases '],marker=dict(color=Italy['Cummulative cases '],coloraxis="coloraxis",)),1,1)
fig.add_trace(go.Bar(x=Brazil['Date '],y=Brazil['Cummulative cases '],marker=dict(color=Brazil['Cummulative cases '],coloraxis="coloraxis",)),1,2)
fig.add_trace(go.Bar(x=India['Date '],y=India['Cummulative cases '],marker=dict(color=India['Cummulative cases '],coloraxis="coloraxis",)),2,1)
fig.add_trace(go.Bar(x=Russia['Date '],y=Russia['Cummulative cases '],marker=dict(color=Russia['Cummulative cases '],coloraxis="coloraxis",)),2,2)
fig.update_layout(coloraxis=dict(colorscale='Bluered_r'),showlegend=False,title_text='Total Cases(Cummulative)')
fig.update_layout(plot_bgcolor='rgb(230,230,230)')
fig.show()
```
